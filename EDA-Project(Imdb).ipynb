#Cleaning and Preprocessing for Analysis

1.title_basics_df Cleaning

```python
# Creating a copy of the DataFrame where titleType is 'movie'
movies_df = title_basics_df[title_basics_df['titleType'] == 'movie'].copy()

# Now you can safely modify movies_df without warnings
movies_df['genres'].fillna('Unknown', inplace=True)
movies_df['startYear'] = pd.to_numeric(movies_df['startYear'], errors='coerce')
movies_df['runtimeMinutes'] = pd.to_numeric(movies_df['runtimeMinutes'], errors='coerce')
movies_df['runtimeMinutes'].fillna(movies_df['runtimeMinutes'].median(), inplace=True)
```

2.title_ratings_df Cleaning
Since title_ratings_df appears clean, no additional steps are required here.

3.name_basics_df Cleaning

```python
name_basics_df['primaryProfession'] = name_basics_df['primaryProfession'].apply(lambda x: x.split(',')[0] if isinstance(x, str) else 'Unknown')
```

Merging Datasets
Merge movies_df with title_ratings_df:

```python
movies_with_ratings_df = pd.merge(movies_df, title_ratings_df, on='tconst', how='left')
```

Additional Preprocessing for Genre Analysis
Split genres into Separate Rows:
This is necessary to analyze individual genres. We can use Pandas' explode function.

```python
movies_with_ratings_df['genres'] = movies_with_ratings_df['genres'].str.split(',')
movies_exploded_by_genre = movies_with_ratings_df.explode('genres')
```

1. Genre Analysis
Objective:
Understand the distribution and popularity of different genres.

Explanation:
This bar chart visualizes the number of movies in each genre. It helps identify the most and least popular genres in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming movies_exploded_by_genre DataFrame is ready
genre_counts = movies_exploded_by_genre['genres'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Popularity of Movie by Genres')
plt.show()
```

💡Insights:
Drama movies are the most common, showing they're really popular, while there aren't as many war or musical movies, which probably means they're for special tastes or are harder to make.

2.Rating Distribution Analysis
Objective:
Explore the distribution of movie ratings.

Explanation:
The histogram shows the distribution of average ratings across movies. It reveals the commonality of ratings and helps identify if there are any specific rating patterns.

```python
# Assuming movies_with_ratings_df DataFrame is ready
plt.figure(figsize=(10, 6))
sns.histplot(movies_with_ratings_df['averageRating'], kde=True, bins=30)
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Ratings')
plt.show()
```

💡Insights:
most movies get scores between 6 and 8 out of 10, with very few movies scoring very low or very high.

3. Temporal Analysis
Objective:
Analyze how the production of movies has evolved over time.

Explanation:
This line graph depicts the number of movies produced each year, providing insight into the evolution and growth of the film industry.

```python
yearly_counts = movies_with_ratings_df['startYear'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies Produced')
plt.title('Movie Production Over the Years')
plt.show()
```

💡Insights:
The graph tells us that over the years, the number of movies being made kept going up slowly, then in recent years there was a really big rise, but after that, it dropped off quite suddenly.

4. Correlation Analysis
Objective:
Investigate relationships between ratings, number of votes, and runtime.

Explanation:
The heatmap displays the correlation between various numerical variables like ratings, number of votes, and runtime, indicating any strong relationships or lack thereof.

```python
correlation_matrix = movies_with_ratings_df[['averageRating', 'numVotes', 'runtimeMinutes']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Rating, Votes, and Runtime')
plt.show()
```

💡Insights:
The correlation matrix shows very weak relationships between movie ratings, the number of votes, and movie runtime. All correlation values are close to zero, indicating that none of these factors strongly predict the others.

To find and visualize the 10 most highly watched movies in 2023 based on the number of votes (numVotes), you'll need to filter the dataset for movies from the year 2023, sort them by numVotes, and then plot the results.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Filter for movies in 2023
movies_2023 = movies_with_ratings_df[movies_with_ratings_df['startYear'] == 2023]

# Sort by 'numVotes' to get the most watched movies
most_watched_movies_2023 = movies_2023.sort_values(by='numVotes', ascending=False).head(10)

# Create the bar plot for the top 10 most watched movies in 2023
plt.figure(figsize=(12, 8))
sns.barplot(x='numVotes', y='primaryTitle', data=most_watched_movies_2023)
plt.title('Top 10 Most Highly Watched Movies of 2023')
plt.xlabel('Number of Votes')
plt.ylabel('Movie Title')
plt.show()
```

💡Insights:
It shows that "Oppenheimer" was the most watched movie in 2023, receiving the highest number of votes, closely followed by "Barbie". This indicates that these movies had a lot of viewers and were very popular. The range of movies includes both action films and family-friendly options, showing that different types of movies were popular in 2023.
